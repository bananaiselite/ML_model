```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
iris = load_iris()
y = iris.target
X = iris.data

X.shape
```

```python
pd.DataFrame(X)
```

```python
pca = PCA(n_components=2) 
X_dr = pca.fit_transform(X)
X_dr.shape
```

components_ are mathematically the eigenvectors of the covariance matrix of the centered input matrix.

This can be verified by using plain numpy.

```python
PCA().fit(X).components_
```

```python
V = PCA(2).fit(X).components_
```

```python
np.dot(X,V.T)
```

```python
np.dot(X,V.T)-X_dr
```

```python
#(Xi-Xbar)*VT = PCA(X)
np.dot(X-np.mean(X, axis=0),V.T)-X_dr
```

# SVD

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
faces.images.shape
```

```python
faces.data.shape
```

```python
fig, axes = plt.subplots(3,8
                        ,figsize=(8,4)
                        ,subplot_kw = {"xticks":[],"yticks":[]} 
                        )
for i, ax in enumerate(axes.flat):
    ax.imshow(faces.images[i,:,:]
              ,cmap="gray" )
```

```python
X = faces.data
pca = PCA(150).fit(X)
V = pca.components_
```

```python
fig, axes = plt.subplots(3,8
                         ,figsize=(8,4)
                         ,subplot_kw = {"xticks":[],"yticks":[]})
for i, ax in enumerate(axes.flat):
    ax.imshow(V[i,:].reshape(62,47),cmap="gray")
```
